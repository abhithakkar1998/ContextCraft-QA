#### Create files and save it to dataset folder

```python
import os
import wikipedia
```

```python
search_terms = ['Artificial Intelligence', 'Machine Learning','Deep Learning', 'Neural Networks', 'Generative AI', 
               'Computer Vision', 'Large Language Model', 'Retrieval-augmented generation', 'Object Detection', 
               'Face Recognition', 'Natural Language Processing', 'Image Processing']  # You can change this to your desired topic
num_results = 10  # You can adjust the number of search results to consider
output_folder = "dataset"
```

```python
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Created folder: {output_folder}")
else:
    print(f"Folder '{output_folder}' already exists.")
```

```python
for i, term in enumerate(search_terms):
    try:
        # Perform the Wikipedia search
        search_results = wikipedia.search(term)
        print(search_results)
        if (search_results != None) and (len(search_results) != 0):
            for j, result in enumerate(search_results):
                print(f"{(i*10)+j+1}. Getting content for {result}")
                try:
                    page = wikipedia.page(result, auto_suggest=False)
                    if page != None:
                        pg_title = page.title #leaving out the title as it is creatinbg noise
                        pg_summary = page.summary
                        #pg_content = page.content #leaving out the content as it includes some unnecessary data which might require cleaning. 
                        filename = f"{pg_title.replace(' ', '_')}.txt"
                                                
                        # Save the formatted content to the text file in the dataset folder
                        filepath = os.path.join(output_folder, filename)
                        if os.path.exists(filepath):
                            print(f"File '{filename}' already exists. Skipping.")
                            continue  # Move to the next search result
                        
                        #file_content = f"Title: {pg_title}\n\nSummary:\n{pg_summary}\n\nContent:\n{pg_content}"
                        file_content = pg_summary
                        with open(filepath, "w", encoding="utf-8") as f:
                            f.write(file_content)
                            print(f"Content for '{pg_title}' saved to '{filepath}'")
                            print("\tDocument Created Successfully!")
                    else:
                        print("\tWikipedia Page Empty")
                except wikipedia.exceptions.PageError:
                    print(f"Error: Page '{result}' not found on Wikipedia (after search).")
                except wikipedia.exceptions.DisambiguationError as e:
                    print(f"Error: '{result}' is ambiguous. Please be more specific. Options: {e.options[:5]}")
                except Exception as e:
                    print(f"An unexpected error occurred while fetching '{result}': {e}")
    except wikipedia.exceptions.WikipediaException as e:
        print(f"An error occurred during the Wikipedia search: {e}")
```
